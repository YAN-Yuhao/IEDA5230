## integer programming
It's obvious that the larger $m$ is, the lower total cost can be. \
If $m=20$, it's a typical TSP. So when $m=6$, we can get the mini-cost. \
I spilt the sequence by 6 subsequence, and run integer programming to solve yhe problem. \
The result shows that the best result is 155.

```python
dis=np.array(E)
for i in range(N):
    for j in range(i+1,N+i):
        dis[i][j % N]=dis[i][(j-1) % N]+E[(j-1) % N][j % N]
```

```python
def solve(b1,b2,b3,b4,b5):
    S=[0,b1,b2,b3,b4,b5]
    F=[b1-1,b2-1,b3-1,b4-1,b5-1,N-1]
    constant=dis[S[0]][F[0]]+dis[S[1]][F[1]]+dis[S[2]][F[2]]+dis[S[3]][F[3]]+dis[S[4]][F[4]]+dis[S[5]][F[5]]
    d=[]
    for i in range(6):
        for j in range(i+1,6):
            d.append(E[F[i]][S[j]])
     
    ProbLP = pulp.LpProblem(sense=pulp.LpMinimize)
    
    x12 = pulp.LpVariable('x12',lowBound=0, upBound=1, cat='Integer')
    x13 = pulp.LpVariable('x13',lowBound=0, upBound=1, cat='Integer')
    x14 = pulp.LpVariable('x14',lowBound=0, upBound=1, cat='Integer')
    x15 = pulp.LpVariable('x15',lowBound=0, upBound=1, cat='Integer')
    x16 = pulp.LpVariable('x16',lowBound=0, upBound=1, cat='Integer')
    x23 = pulp.LpVariable('x23',lowBound=0, upBound=1, cat='Integer')
    x24 = pulp.LpVariable('x24',lowBound=0, upBound=1, cat='Integer')
    x25 = pulp.LpVariable('x25',lowBound=0, upBound=1, cat='Integer')
    x26 = pulp.LpVariable('x26',lowBound=0, upBound=1, cat='Integer')
    x34 = pulp.LpVariable('x34',lowBound=0, upBound=1, cat='Integer')
    x35 = pulp.LpVariable('x35',lowBound=0, upBound=1, cat='Integer')
    x36 = pulp.LpVariable('x36',lowBound=0, upBound=1, cat='Integer')
    x45 = pulp.LpVariable('x45',lowBound=0, upBound=1, cat='Integer')
    x46 = pulp.LpVariable('x46',lowBound=0, upBound=1, cat='Integer')
    x56 = pulp.LpVariable('x56',lowBound=0, upBound=1, cat='Integer')
    
    u1 = pulp.LpVariable('u1',lowBound=0, cat='Integer')
    u2 = pulp.LpVariable('u2',lowBound=0, cat='Integer')
    u3 = pulp.LpVariable('u3',lowBound=0, cat='Integer')
    u4 = pulp.LpVariable('u4',lowBound=0, cat='Integer')
    u5 = pulp.LpVariable('u5',lowBound=0, cat='Integer')
    u6 = pulp.LpVariable('u6',lowBound=0, cat='Integer')
    
    ProbLP += (d[0]*x12+d[1]*x13+d[2]*x14+d[3]*x15+d[4]*x16+d[5]*x23+d[6]*x24+d[7]*x25+d[8]*x26+d[9]*x34+d[10]*x35+d[11]*x36+d[12]*x45+d[13]*x46+d[14]*x56)  
    
    ProbLP += (u1 -  u2 + 6*x12 <= 5)
    ProbLP += (u1 -  u3 + 6*x13 <= 5)
    ProbLP += (u1 -  u4 + 6*x14 <= 5)
    ProbLP += (u1 -  u5 + 6*x15 <= 5)
    ProbLP += (u1 -  u6 + 6*x16 <= 5)
    ProbLP += (u2 -  u3 + 6*x23 <= 5)
    ProbLP += (u2 -  u4 + 6*x24 <= 5)
    ProbLP += (u2 -  u5 + 6*x25 <= 5)
    ProbLP += (u2 -  u6 + 6*x26 <= 5)
    ProbLP += (u3 -  u4 + 6*x34 <= 5)
    ProbLP += (u3 -  u5 + 6*x35 <= 5)
    ProbLP += (u3 -  u6 + 6*x36 <= 5)
    ProbLP += (u4 -  u5 + 6*x45 <= 5)
    ProbLP += (u4 -  u6 + 6*x46 <= 5)
    ProbLP += (u5 -  u6 + 6*x56 <= 5)
    
    ProbLP += (x12+x13+x14+x15+x16==2)
    ProbLP += (x12+x23+x24+x25+x26==2)
    ProbLP += (x13+x23+x34+x35+x36==2)
    ProbLP += (x14+x24+x34+x45+x46==2)
    ProbLP += (x15+x25+x35+x45+x56==2)
    ProbLP += (x16+x26+x36+x46+x56==2)
    ProbLP.solve()
    return pulp.value(ProbLP.objective)+constant
                    
```

## dynamic programming 
The state transition equation is 
$$dp[i] = \min_{j\in\{i+1,\cdots,i+4\}}(dp[j]+distance[i][j])$$
where the $dp[i]$ is the mini-cost from node i to node 0, and $distance[i][j]$ shows the mini-cost from node i to node j.\
The result shows that the best result is 159.

```python
def Sumdis(idx):
    lenth=len(idx)
    res=0
    for i in range(lenth-1):
        res=res+E[idx[i]% N][(idx[i+1])% N]
    return res
    
dist=np.ones((N, N)) * np.inf  

for i in range(N):
    dist[i][(i+1) % N]=E[i][(i+1) % N]
    dist[i][(i+2) % N]=dist[i][(i+1) % N]+E[(i+1) % N][(i+2) % N]
    dist[i][(i+3) % N]=min(dist[i][(i+2) % N]+E[(i+2) % N][(i+3) % N], Sumdis([i,i+2,i+1,i+3]))
    dist[i][(i+4) % N]=min(dist[i][(i+3) % N]+E[(i+2) % N][(i+3) % N], Sumdis([i,i+1,i+3,i+2,i+4]),Sumdis([i,i+3,i+1,i+2,i+4]),Sumdis([i,i+3,i+2,i+1,i+4]),Sumdis([i,i+2,i+3,i+1,i+4]))
```

```python
dp=np.ones(N) * np.inf
dp[N-4:N]=dist[N-4:N,0]
for i in range(5,N+1):
    dp[N-i]=min(dp[N-(i-1)]+dist[N-i][N-(i-1)],dp[N-(i-2)]+dist[N-i][N-(i-2)],dp[N-(i-3)]+dist[N-i][N-(i-3)],dp[N-(i-4)]+dist[N-i][N-(i-4)])
print(dp[0])
```
